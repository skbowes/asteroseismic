```python
def plotter(m,fov,v,harmonic):
    s = pg.read_output('../mesafolder/work/LOGS/M'+str(m)+'_alpha'+str(fov)+'_v'+str(v)+'/pmodefrequencies1_split.txt').group_by(str(harmonic))
    plt.plot(s.groups[0]['n_pg'], s.groups[0]['freq'].real, ".-", label=str(harmonic)+'=1')
    plt.plot(s.groups[1]['n_pg'], s.groups[1]['freq'].real, ".-", label=str(harmonic)+'=2')
#     plt.plot(s.groups[2]['n_pg'], s.groups[2]['freq'].real, ".-", label=str(harmonic)+'=3')
    plt.xlabel('n_pg (radial order)')
    plt.ylabel('Frequency (cyc/day)')
    plt.legend()
    plt.show()
# m = [1.4-2.1]
# fov = [0,0.01,0.02,0.03,0.04]
# v = [200,215,230,245,260,275,290,300]
```
